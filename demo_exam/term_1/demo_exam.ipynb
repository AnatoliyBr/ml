```python
import pandas as pd
import numpy as np

df = pd.read_csv('/report.csv')
df.head()
```

```python
normalized_df = (df - df.min()) / (df.max() - df.min())
normalized_df.head()
```

```python
round(normalized_df.MIP.mean(), 3)
```

```python
X_train = normalized_df.drop(columns='TARGET')
y_train = normalized_df.TARGET
```

```python
from sklearn.linear_model import LogisticRegression


log_reg_model = LogisticRegression(random_state = 2019, solver = 'lbfgs').fit(X_train, y_train)
```

```python
x_test  = np.array([0.237, 0.684, 0.234, 0.169, 0.149, 0.674, 0.462, 0.82])

np.round(log_reg_model.predict_proba([x_test]), 3)
```
