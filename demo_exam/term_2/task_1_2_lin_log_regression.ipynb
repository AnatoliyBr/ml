## Задание 1

Уравнение линейной регрессии задается соотношением: $2 + 3X_1 - 4X_2 = 0$

Предскажите отклик  для тестового объекта $X_2$ — (4):

Ответ: $X_1 = 3.5$

Может ли данное уравнение задавать регрессию, построенную по тренировочным данным: (1, 2) и (-1, 0)? Если нет, постройте правильное уравнение и введите его в поле ввода, если да — введите 1.

$2 + \theta_1 + \theta_2 2 = 0$  
$2 - \theta_1 = 0$  
$=> \theta_1 = 2 => \theta_2 = -2$

Ответ: $X_1 - X_2 + 1$

```python
import numpy as np

theta = np.array([[1, 2], [-1, 0]])
y = np.array([-2, -2])
x = np.linalg.solve(theta, y)

print(f'{x}\n{np.allclose(np.dot(theta, x), y)}')
```

## Задание 2

Модель логистической регрессии обучена на определение класса +.

$\Psi = -2.992 + 0.240 X_1 - 1.102 X_2$

На рисунке отображены точки, на которых обучена модель, но вам неизвестно какие из них относятся к классу + и -. 

Учитывая, что тестовая точка A с координатами (0, 0) относится к классу +, верно ли она классифицирована моделью?

```python
psi = -2.992 + 0.240 * 0 - 1.102 * 0
psi
```

```python
from numpy import exp

P = round(1 / (1 + exp(-psi)), 3)
P
```
