```python
import pandas as pd


df = pd.read_csv('report.csv')
df_norm = (df - df.min()) / (df.max() - df.min())
df_norm.head()
```

```python
X_train = df_norm.drop(columns='TARGET')
X_train.head()
```

```python
Y_train = df_norm.TARGET
Y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression


log_reg_model = LogisticRegression(random_state=2019, solver='lbfgs').fit(X_train, Y_train)

test = np.array([0.237, 0.293, 0.897, 0.0, 0.841, 0.969, 0.138, 0.083])

log_reg_model.predict_proba([test])
```

```python
X_train.iloc[:,:].iterrows()
```

```python
from scipy.spatial import distance

min_distance = np.min([distance.euclidean(test, obj) for id, obj in X_train.iloc[:,:].iterrows()])
round(min_distance, 3)
```
