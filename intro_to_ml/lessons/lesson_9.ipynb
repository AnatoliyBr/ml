```python
from numpy import exp

theta_0 = -0.046
theta_1 = 0.541
theta_2 = -0.014
theta_3 = -0.132

x_1 = 4
x_2 = 40
x_3 = 8

P = 1 / (1 + exp(-(theta_0 + theta_1 * x_1 + theta_2 * x_2 + theta_3 * x_3)))
round(P, 3)
```

```python
import numpy as np
from scipy.stats import binomtest

k=8
n=10
p=0.8

res = binomtest(k, n, p)
res.proportion_estimate
```
