```python
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances, manhattan_distances, pairwise_distances
import io
import numpy as np


x = [(1, 3, -5, 7)]
x_ = [(2, -4, 0, 8)]
```

```python
euclidean_distances(x, x_)
```

```python
manhattan_distances(x, x_)
```

```python
pairwise_distances(x, x_, metric='chebyshev')
```

```python
data = '''id,X,Y,Class
1,28,10,1
2,49,49,1
3,48,35,0
4,36,33,1
5,45,54,0'''

data = pd.read_csv(io.StringIO(data), index_col='id')
data
```

```python
df = list(zip(data.X, data.Y))
df
```

```python
euclidean_distances(df, [(33, 47)])
```

```python
manhattan_distances(df, [(33, 47)])
```

```python
pairwise_distances(df, [(33, 47)], metric='chebyshev')
```

```python
class_0 = data[data.Class == 0]
class_0 = list(zip(class_0.X, class_0.Y)) 
```

```python
w_sum_0 = sum(1 / (euclidean_distances(class_0, [(33, 47)])) ** 2)
np.round(w_sum_0, 3)
```

```python
class_1 = data[data.Class == 1]
class_1 = list(zip(class_1.X, class_1.Y)) 
```

```python
w_sum_1 = sum(1 / (euclidean_distances(class_1, [(33, 47)])) ** 2)
np.round(w_sum_1, 4)
```
