## Упражнение 5.1

Ниже приведена схема марковского процесса принятия решений.

Для всех состояний известны распределения вероятностей выбора доступных действий $\pi(a|s)$, распределения вероятностей переходов в следующие состояния $P_{ss'}^a$ и математические ожидания $R_{ss'}^a$ получаемых при этом наград.

Запишите уравнения Беллмана и получите значения ценности состояний, если коэффициент дисконтирования $\gamma = 0.8$.

```python
from sympy import symbols, solve

v1, v2, v3, v4 = symbols('v1 v2 v3 v4')

gamma = 0.8

eq1 = 0.6 * (0.3 * (2.0 + gamma * v2) + 0.7 * (3.0 + gamma * v3)) + 0.4 * (3.0 + gamma * v3) - v1
eq2 = 3.0 + gamma * v1 - v2
eq3 = 0.3 * (-3.0 + gamma * v1) + 0.7 * (0.2 * (1.0 + gamma * v3) + 0.8 * (6.0 + gamma * v4)) - v3
eq4 = 0.6 * (5.0 + gamma * v1) + 0.4 * (-3.0 + gamma * v2) - v4
```

```python
eq1
```

```python
solution = solve((eq1, eq2, eq3, eq4), (v1, v2, v3, v4))
```

```python
solution
```

```python
list([round(value, 3) for key, value in solution.items()])
```
